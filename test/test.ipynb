```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
```

```python
data = pd.read_csv('../homework3/encodings/faces_all_simulation.txt')

X = data.iloc[:, :-1].values
y = data.iloc[:, 128].values
class_names = np.unique(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

#classifier = svm.SVC(kernel='linear', C=0.01).fit(X_train, y_train)
classifier = GaussianNB().fit(X_train, y_train)

disp = plot_confusion_matrix(classifier, X_test, y_test, display_labels=class_names, xticks_rotation='vertical', cmap=plt.cm.Blues)
#print(disp.confusion_matrix)
plt.show(disp)
```
